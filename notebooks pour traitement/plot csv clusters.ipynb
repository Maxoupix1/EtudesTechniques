```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les donnes
file_path = "cluster_distribution.csv"
df = pd.read_csv(file_path)

# Dfinir une palette de couleurs personnalise
palette = sns.color_palette("coolwarm", n_colors=df["Cluster"].nunique())

# Cration du graphique
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x="Time Step", y="Dyad Count", hue="Cluster", marker="o", palette=palette)

# Personnalisation
title = "volution du nombre de dyades par cluster"
plt.title(title)
plt.xlabel("Time Step")
plt.ylabel("Dyad Count")
plt.legend(title="Cluster")
plt.grid(True)

# Afficher le graphique
plt.show()
```
