```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les donnes
file_path = "cluster_distribution diag 120 6.csv"
df = pd.read_csv(file_path)

# Obtenir les clusters uniques
clusters = df["Cluster"].unique()
num_clusters = len(clusters)

# Crer une figure avec des sous-graphiques plus compacts
fig, axes = plt.subplots(nrows=num_clusters, figsize=(10, 2 * num_clusters), sharex=True)

# S'assurer que axes est une liste si un seul cluster
if num_clusters == 1:
    axes = [axes]

# Dfinir une palette de couleurs
palette = sns.color_palette("coolwarm", num_clusters)

# Gnrer un graphique par cluster
for ax, cluster, color in zip(axes, clusters, palette):
    cluster_df = df[df["Cluster"] == cluster]
    sns.lineplot(data=cluster_df, x="Time Step", y="Dyad Count", marker="o", color=color, ax=ax)
    
    ax.set_title(f"Cluster {cluster}", fontsize=10)
    ax.set_ylabel("Dyad Count", fontsize=9)
    ax.grid(True)

# Ajuster l'espacement et cacher les labels inutiles
plt.xlabel("Time Step", fontsize=10)
plt.tight_layout()
plt.show()
```
